```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_data = """Month,Store,Sales
Jan,Store A,12000
Jan,Store B,15000
Feb,Store A,18000
Feb,Store B,17000
Mar,Store A,16000
Mar,Store B,19000
"""

with open('sales_data.csv', 'w') as file:
    file.write(csv_data)
```

```python
df = pd.read_csv('sales_data.csv')
print(df)
```

```python
sales = df.groupby('Store')['Sales'].sum()
print(sales)
```

```python
ts = sales.idxmax()
print(ts)
```

```python
ms = df.groupby('Month')['Sales'].sum()
print(ms)
```

```python
lchart = df.pivot(index='Month', columns='Store', values='Sales')
lchart.plot(kind='line', marker='o')
plt.title("Monthly Sales by Store")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(sales.index, sales.values, color=['blue', 'green'])
plt.title("Total Sales per Store")
plt.ylabel("Sales")
plt.grid(True)
plt.savefig('sales chart.png')
plt.show()
```

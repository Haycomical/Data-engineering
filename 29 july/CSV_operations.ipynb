```python
import pandas as pd
import matplotlib.pyplot as plt

csv_data = """Name,Math,Science,English
Amit,78,85,74
Priya,92,89,96
Kiran,65,70,60
Anita,88,95,90
Ravi,55,60,58
"""

with open('students.csv', 'w') as f:
    f.write(csv_data)


df = pd.read_csv('students.csv')
print("First 3 rows:")
print(df.head(3))
```

```python
df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)
```

```python
print(df)
```

```python
print("\nHighest average:")
print(df.loc[df['Average'].idxmax()])
```

```python
print(df[df['Science']>80])
```

```python
print(df[df['Science'] <70])
```

```python
print(df.sort_values(by='English',ascending=False).head(3))
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(8,5))
plt.bar(df['Name'], df['Average'], color='skyblue')
plt.title('Average Scores')
plt.xlabel('Name')
plt.ylabel('Average Score')
plt.show()
```

```python
df.set_index('Name')[['Math','Science','English']].plot(marker='*')
plt.title('Scores in Each Subject')
plt.ylabel('Scores')
plt.show()
```
